```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Build Neural Model

```python
# import libraries
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, InputLayer
from tensorflow.keras.optimizers import Adam

from sklearn.metrics import confusion_matrix
```

```python
x_input = x_train.shape[1:]
```

```python
model = Sequential()
model.add(InputLayer(x_input))
model.add(Dense(10, activation='relu', kernel_initializer='uniform'))
model.add(Dense(10, activation='relu', kernel_initializer='uniform'))
model.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics='accuracy')
```

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=8, validation_split=0.1)
```

```python
hist = pd.DataFrame(history.history)
hist
```

```python
hist.plot()
```

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy')-logs.get('val_accuracy')) > 0.01: 
            print('Validation Accuracy fell below 1% of training accuracy. Cancelling Training!')
            self.model.stop_training = True
        
callbacks = myCallback()

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics='accuracy')

history = model.fit(x_train, y_train, epochs=20, batch_size=8, validation_split=0.1, callbacks=[callbacks])
```

# y_pred

```python
y_pred = model.predict(x_test)
y_pred = (y_pred>0.5).astype(int)
# y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy = (cm[0][0] + cm[1][1])/(cm[0][1] + cm[1][0] + cm[0][0] + cm[1][1])
```

```python
accuracy
```

```python
print('The accuracy of the model is: ', accuracy)
```
